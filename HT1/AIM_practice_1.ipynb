```python
import pandas

data = pandas.read_csv('titanic.csv')

print(data)
```

```python
female = data[data['Sex'] == 'female'].dropna()
male = data[data['Sex'] == 'male'].dropna()

print(female)
print(male)
```

```python
import matplotlib.pyplot as plt

data['Age'].plot(kind='hist', bins=100, alpha=0.7)

plt.xlabel('Возраст')
plt.ylabel('Количество')

plt.grid(axis = 'y')
```

```python
print(male['Age'].min())
print(male['Age'].max())
print(male['Age'].mean())
print(male['Age'].mode()[0])
print(male['Age'].median())
```

```python
print(female['Age'].min())
print(female['Age'].max())
print(female['Age'].mean())
print(female['Age'].mode()[0])
print(female['Age'].median())
```

```python
female['Age'].plot(kind='line', label='Male', color='blue', linestyle='-')
male['Age'].plot(kind='line', label='Female', color='pink', linestyle='-')

plt.title('kind=line')
plt.ylabel('Age')
plt.legend()

plt.show()
```

```python
male['Age'].plot(kind='hist', alpha=0.5, label='Male', color='blue', bins=10)
female['Age'].plot(kind='hist', alpha=0.5, label='Female', color='pink', bins=10)

plt.title('kind=hist')
plt.xlabel('Возраст')
plt.ylabel('Частота')
plt.legend()
plt.show()
```

```python
plt.scatter(male.index, male['Age'], label='male', color='blue', alpha=0.6)
plt.scatter(female.index, female['Age'], label='female', color='pink', alpha=0.6)

plt.title('scatter')
plt.ylabel('Возраст')
plt.legend()
plt.show()
```
